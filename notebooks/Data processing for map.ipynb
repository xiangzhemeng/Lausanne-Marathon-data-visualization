```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/marathon_data.csv')
```

```python
df.head()
```

```python
df['rank'] = df['rank'].apply(lambda x: x.replace('.',''))
```

```python
df['time by secs'] = df['time'].apply(lambda x: int(x.split(":")[0])*3600 + int(x.split(":")[1])*60 + float(x.split(":")[2]))
```

```python
df_new = df.groupby(['race year','age group','gender','category']).apply(lambda x: x.sort_values(['time by secs']))
```

```python
df_new.drop(['race year','age group','gender','category'],axis=1,inplace=True)
df_top10 = df_new.groupby(['race year','age group','gender','category']).head(10)
```

```python
df_top10 = df_top10.reset_index()
```

```python
df_top10 = df_top10[['city','canton','country','latitude','longitude']]
```

```python
df_top10.to_csv('marethon_top10.csv')
```

```python
df_participant_count = df.groupby(['race year','age group','gender','city','latitude','longitude']).count()
```

```python
df_participant_count = df_participant_count.reset_index()
```

```python
df_participant_count.drop(['birth year','age','category','rank','time','canton','country','time by secs'],axis=1,inplace=True)
```

```python
df_dd = df_participant_count.groupby(['race year','age group','gender']).apply(lambda x: x.sort_values(['name'], ascending=False))
```

```python
df_dd.drop(['race year','age group','gender'],axis=1,inplace=True)
```

```python
df_dd=df_dd.reset_index()
df_dd
```

```python
df_dd.drop(['level_3'],axis=1,inplace=True)
```

```python
df_city_counts_top20 = df_dd.groupby(['race year','age group','gender','city','latitude','longitude']).head(20)
```

```python
df_city_counts_top20 = df_city_counts_top20.set_index(['race year','age group','gender','city','latitude','longitude'])
```

```python
df_city_counts_top20.to_csv('df_city_counts_top20.csv')
```
